```python
import pandas as pd
import torch
import ast

# ÐÑÐµÐ´Ð¿Ð¾Ð»Ð°Ð³Ð°ÐµÐ¼, ÑÑÐ¾ df2 ÑÐ¶Ðµ Ð·Ð°Ð³ÑÑÐ¶ÐµÐ½ Ð¸ ÑÐ¾Ð´ÐµÑÐ¶Ð¸Ñ ÑÑÐ¾Ð»Ð±ÐµÑ 'tokens'

# Ð¤ÑÐ½ÐºÑÐ¸Ñ Ð´Ð»Ñ Ð¾Ð±ÑÐ°Ð±Ð¾ÑÐºÐ¸ ÑÐ¾ÐºÐµÐ½Ð¾Ð²
def process_tokens(tokens_str):
    # ÐÐ°Ð¼ÐµÐ½ÑÐµÐ¼ 'tensor' Ð½Ð° 'torch.tensor' Ð¸ ÑÐ°Ð·Ð±Ð¸ÑÐ°ÐµÐ¼ ÑÑÑÐ¾ÐºÑ
    try:
        tokens_dict = ast.literal_eval(tokens_str.replace("tensor", "torch.tensor"))
        input_ids_tensor = torch.tensor(tokens_dict['input_ids'])
        attention_mask_tensor = torch.tensor(tokens_dict['attention_mask'])
        return input_ids_tensor, attention_mask_tensor
    except Exception as e:
        print(f"Error processing tokens: {e}")
        return None, None

# ÐÐ±ÑÐ°Ð±Ð¾ÑÐºÐ° Ð²ÑÐµÑ ÑÐ¾ÐºÐµÐ½Ð¾Ð² Ð² df2
input_ids_list = []
attention_masks_list = []

for tokens_str in df2['tokens']:
    input_ids_tensor, attention_mask_tensor = process_tokens(tokens_str)
    input_ids_list.append(input_ids_tensor)
    attention_masks_list.append(attention_mask_tensor)

# ÐÐ¾Ð±Ð°Ð²Ð»ÑÐµÐ¼ Ð½Ð¾Ð²ÑÐµ ÐºÐ¾Ð»Ð¾Ð½ÐºÐ¸ Ð² DataFrame
df2['input_ids'] = input_ids_list
df2['attention_mask'] = attention_masks_list

# Ð¡Ð¾ÑÑÐ°Ð½ÑÐµÐ¼ Ð¾Ð±Ð½Ð¾Ð²Ð»ÐµÐ½Ð½ÑÐ¹ DataFrame Ð² CSV (Ð¸Ð»Ð¸ Ð² Ð´ÑÑÐ³Ð¾Ð¹ ÑÐ¾ÑÐ¼Ð°Ñ Ð¿Ð¾ Ð²Ð°ÑÐµÐ¼Ñ Ð²ÑÐ±Ð¾ÑÑ)
df2.to_csv('updated_data.csv', index=False)

print("Ð¢Ð¾ÐºÐµÐ½Ñ ÑÑÐ¿ÐµÑÐ½Ð¾ Ð¾Ð±ÑÐ°Ð±Ð¾ÑÐ°Ð½Ñ Ð¸ ÑÐ¾ÑÑÐ°Ð½ÐµÐ½Ñ Ð² 'updated_data.csv'.")
```
